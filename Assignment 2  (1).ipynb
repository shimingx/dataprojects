```python
import requests
import json
import pandas as pd
```

```python
url= ('http://www.patentsview.org/api/patents/query?q={"patent_year":"2010"}&f=["patent_title","patent_date","patent_type","patent_processing_time","app_date","inventor_first_name","inventor_last_name","inventor_state","assignee_first_name","assignee_last_name","assignee_organization","assignee_state","assignee_type"]&s=[{"patent_title":"asc"}]')
```

```python
print(url)
```

```python
r = requests.get(url)
json = r.json()
```

```python
r.status_code
print (r.content)
```
